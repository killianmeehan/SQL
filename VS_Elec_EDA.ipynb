```python
#converting invoice time to a timestamp and date to datetime obj
t= invoices['invoice_time']
invoices['invoice_time'] = pd.to_datetime(t, format='%H').dt.time

#d = invoices['invoice_date']
#invoices['invoice_date'] = pd.to_datetime(d).dt.date

invoices['invoice_date'] = invoices.apply(lambda r : pd.datetime.combine(r['invoice_date'],r['invoice_time']),1)

invoices.drop('invoice_time', axis=1, inplace=True)
```
